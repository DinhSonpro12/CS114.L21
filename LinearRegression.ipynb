```python
# Thêm các thư viện cần thiết
import os
import numpy as np
import pandas as pd

# Đường dẫn thư mục dataset
data_2 = pd.read_csv("https://cs.uit.edu.vn/data2.txt")
data_2.columns = ['Size','Time']
data_2.to_csv('data2.csv', 
                  index = None)
data_3 = pd.read_csv("https://cs.uit.edu.vn/data3.txt")
data_3.columns = ['Size','Time']
data_3.to_csv('data3.csv', 
                  index = None)
data_4 = pd.read_csv("https://cs.uit.edu.vn/data4.txt")
data_4.columns = ['Size','Time']
data_4.to_csv('data4.csv', 
                  index = None)
# data2.csv
x_2 = np.array(data_2['Size']).reshape(-1, 1)
y_2 = np.array(data_2['Time'])

# data3.csv
x_3 = np.array(data_3['Size']).reshape(-1, 1)
y_3 = np.array(data_3['Time'])

# data4.csv
x_4 = np.array(data_4['Size']).reshape(-1, 1)
y_4 = np.array(data_4['Time'])
```

```python

# Thêm thư viện cần thiết
import numpy as np
from math import sqrt
from sklearn.metrics import mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt

# Hàm in kết quả
def printResult(y, y_predict):
  data_mse = mean_squared_error(y, y_predict)
  data_mae = mean_absolute_error(y, y_predict)
  print("Mean squared error: ", data_mse)
  print("Mean absolute error: ", data_mae)
  

# Hàm biểu diễn kết quả
def showResult(x, y, y_predict, title):
  plt.figure(figsize=(5, 5))
  plt.plot(x, y, 'b.')
  plt.plot(x, y_predict, 'r-')
  plt.xlabel('Size')
  plt.ylabel('Time')
  plt.title(title)
  plt.show
```

```python

from sklearn.linear_model import LinearRegression
lR_2 = LinearRegression().fit(x_2, y_2)

print("===== data2.csv =====")
y_2_predict = lR_2.predict(x_2)
showResult(x_2, y_2, y_2_predict, 'data2.csv')
printResult(y_2,y_2_predict)

print("===== data3.csv =====")
y_3_predict = lR_2.predict(x_3)
showResult(x_3, y_3, y_3_predict, 'data3.csv')
printResult(y_3,y_3_predict)

print("===== data4.csv =====")
y_4_predict = lR_2.predict(x_4)
showResult(x_4, y_4, y_4_predict, 'data4.csv')
printResult(y_4,y_4_predict)
```
